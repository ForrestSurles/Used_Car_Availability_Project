```python
for k,v in letter_dict.items():
    print(f'Key: {k}\nValue: {v}\n')
```

```python
# print(f'Columns: {vehicles_used_pd.columns.values}')
region_list = list(vehicles_used_pd['region'].unique())
entry_list = {}
start_here = '\A'
for k, v in letter_dict.items():
    r = re.compile(f'{start_here}[{v[0]}{v[1]}]')
    entry_list[k] = list(filter(r.match,region_list))

```

```python
letters = ['f', 'g']
print(json.dumps(list((entry_list.get(x) for x in letters)), indent=0))
```

## <font color=MediumSlateBlue>Analyses</font>

```python
# FIRST CODE EXAMPLE: ODOMETER
region_1 = 'houston'
region_2 = 'san diego'
odo_low = 75000
odo_high = 100000
index = ['region','condition', 'manufacturer']
return_columns = ['type', 'year', 'price', 'odometer', 'paint_color', 'id', 'lat', 'long']

a1_df = vehicles_used_pd.loc[
    (vehicles_used_pd['region'] == region_1) | 
    (vehicles_used_pd['region'] == region_2)
]
a2_df = a1_df.loc[
    (a1_df['odometer'] >= odo_low) & 
    (a1_df['odometer'] < odo_high)
]
a2_df = a2_df.set_index(index)
a2_df[return_columns]
```

```python
# with Pablo:
pablo_df = a2_df.copy()
```

```python
pablo_df
```

```python
pablo_table = pablo_df.reset_index()
pablo_table = pablo_table.groupby(['region','type', 'manufacturer'])['type'].count()

    # Number of Vehicles
    # 
pablo_table.head()
```

```python
pablo_table.hvplot.bar(x='type', y='manufacturer', by='region')
```

```python
# SECOND CODE EXAMPLE: PRICE
region_1 = 'houston'
region_2 = 'san diego'
price_low = 15000
price_high = 30000
index = ['region', 'condition', 'manufacturer']
return_columns = ['type', 'year', 'price', 'odometer', 'paint_color', 'id', 'lat', 'long']

b1_df = vehicles_used_pd.loc[
    (vehicles_used_pd['region'] == region_1) | 
    (vehicles_used_pd['region'] == region_2)
]
b2_df = b1_df.loc[
    (b1_df['price'] >= price_low) & 
    (b1_df['price'] < odo_high)
]
b2_df = b2_df.set_index(index)
b2_df[return_columns]
```
