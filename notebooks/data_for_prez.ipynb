# Delete All Incomplete Entries

## <font color=MediumSlateBlue>1) Imports</font>

```python
import pandas as pd
from pathlib import Path
```

```python
full_df = pd.read_csv(Path('./Resources/raw_data_combined_v2.csv'), index_col='uid')
```

```python
print(f'Column Names:\n\n{full_df.columns.values}\n')
print(f'Column Types:\n\n{full_df.dtypes}\n')
print(f'Review Subset Data:\n\n{full_df}')
```

## <font color=MediumSlateBlue>2) Identify & Drop Null Entries</font>

### <font color=PaleGreen>2a) Count Number of Entries Before Dropping</font>

```python
full_df.isnull().sum()
```

### <font color=PaleGreen>2b) Drop Null Entries from the DataFrame</font>

```python
# drop null entries:
# year/manufacturer/odometer(?)/lat/long/posting_date
full_df = full_df.dropna(subset=['posting_date', 'lat', 'long', 'manufacturer', 'year'])
```

### <font color=PaleGreen>2c) Verify removed entries</font>

```python
full_df.isnull().sum()
```

## <font color=MediumSlateBlue>3) Save Filtered DataFrame to a CSV</font>

### <font color=PaleGreen>3a) Save to a CSV</font>

```python
full_df.to_csv(Path('./Resources/data_for_prez.csv'))
```
